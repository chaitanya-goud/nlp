```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/imdb_top_1000.csv')

# Displaying the content and top 5 rows!!
df.head()
```

```python
print("Dataset Info:")
df.info()

print("\nDataset Description (including all columns):")
print(df.describe(include='all'))

unique_counts = {
    "Series_Title": df["Series_Title"].nunique(),
    "Released_Year": df["Released_Year"].nunique(),
    "Certificate": df["Certificate"].nunique(),
    "Runtime": df["Runtime"].nunique(),
    "Genre": df["Genre"].nunique(),
    "IMDB_Rating Mean": df["IMDB_Rating"].mean(),
    "Meta_score Mean": df["Meta_score"].mean(),
    "Director": df["Director"].nunique(),
    "Star1": df["Star1"].value_counts().idxmax(),
    "Star2": df["Star2"].value_counts().idxmax(),
    "Star3": df["Star3"].value_counts().idxmax(),
    "Star4": df["Star4"].value_counts().idxmax(),
    "No_of_Votes Mean": df["No_of_Votes"].mean(),
    "Gross Unique Counts": df["Gross"].nunique()
}

print("\nUnique Counts and Additional Insights:")
for key, value in unique_counts.items():
    print(f"{key}: {value}")
```
